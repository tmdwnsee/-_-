- 'Publishers' 열 또한, bs4로 누락값을 채운다.

```python
df_dev = df_game[df_game['Developers'].isna()]
df_pub = df_game[df_game['Publishers'].isna()]
```

```python
import requests # 파이썬 기본 패키지
from bs4 import BeautifulSoup      # bs4 로드

def find_dev(name):
    try:
        name = name.lower().replace(' ', '-')
        name = name.replace(':', '')
        url = f"https://www.metacritic.com/game/{name}/"

        headers = {'User-Agent': 'Mozilla/5.0'}
        html = requests.get(url, headers=headers)
        # url에 정보를 요청
        soup = BeautifulSoup(html.text, 'html.parser') # bs4 파씽

        # html 확인/ 필요한 정보가 들어있는 태그 추출
        dev = soup.find_all('li','c-gameDetails_listItem g-color-gray70 u-inline-block')

        for i in dev[1]:
            developer = i.text.strip('\n          ')
            developer = developer.strip()
            return developer
        else:
            return '0'
    except:
        return '0'
```

```python
import requests # 파이썬 기본 패키지
from bs4 import BeautifulSoup      # bs4 로드

def find_pub(name):
    try:
        name = name.lower().replace(' ', '-')
        name = name.replace(':', '')
        url = f"https://www.metacritic.com/game/{name}/"

        headers = {'User-Agent': 'Mozilla/5.0'}
        html = requests.get(url, headers=headers)
        # url에 정보를 요청
        soup = BeautifulSoup(html.text, 'html.parser') # bs4 파씽

        # html 확인/ 필요한 정보가 들어있는 태그 추출
        pub = soup.find_all('span','g-outer-spacing-left-medium-fluid g-color-gray70 u-block')

        for i in pub[1]:
            return i.text.strip()
        else:
            return '0'
    except:
        return '0'
```

```python
df_dev['Developers'] = df_dev['Name'].apply(find_dev)
df_pub['Publishers'] = df_pub['Name'].apply(find_pub)
```

```python
df_game.loc[df_game['Developers'].isna(), 'Developers'] = df_dev['Developers']
df_game.loc[df_game['Publishers'].isna(), 'Publishers'] = df_pub['Publishers']
```

```python
def find_genre(id):
    try:
        url = f"https://store.steampowered.com/app/{id}/"

        headers = {'User-Agent': 'Mozilla/5.0'}
        html = requests.get(url, headers=headers)
        # url에 정보를 요청
        soup = BeautifulSoup(html.text, 'html.parser') # bs4 파씽

        # html 확인/ 필요한 정보가 들어있는 태그 추출
        genre = soup.find_all('div','details_block')

        for i in genre:
            listA = i.text.split('\n')
            break

        for j in listA:
            if 'Genre' in j:
                return j.split('Genre: ')[1]
        else:
            return '0'
    except:
        return '0'
```

```python
def find_cat(id):
    try:
        url = f"https://store.steampowered.com/app/{id}/"

        headers = {'User-Agent': 'Mozilla/5.0'}
        html = requests.get(url, headers=headers)
        # url에 정보를 요청
        soup = BeautifulSoup(html.text, 'html.parser') # bs4 파씽

        # html 확인/ 필요한 정보가 들어있는 태그 추출
        category = soup.find_all('div','label')
        category_list = []

        for i in category:
            category_list.append(i)

        if len(category_list) != 0:
            return ', '.join(category_list)
        else:
            return '0'
    except:
        return '0'
```

```python
import requests # 파이썬 기본 패키지
from bs4 import BeautifulSoup      # bs4 로드

def find_score(name):
    try:
        name = name.lower().replace(' ', '-')
        name = name.replace(':', '')
        url = f"https://www.metacritic.com/game/{name}/"

        headers = {'User-Agent': 'Mozilla/5.0'}
        html = requests.get(url, headers=headers)
        # url에 정보를 요청
        soup = BeautifulSoup(html.text, 'html.parser') # bs4 파씽

        # html 확인/ 필요한 정보가 들어있는 태그 추출
        score = soup.find_all('div','c-siteReviewScore_background c-siteReviewScore_background-critic_medium')

        if score[0].text != 'tbd':
            return int(score[0].text)
        else:
            return 0
    except:
        return 0
```

```python
import requests # 파이썬 기본 패키지
from bs4 import BeautifulSoup      # bs4 로드

def find_positive(id):
    try:
        url = f"https://store.steampowered.com/app/{id}/"

        headers = {'User-Agent': 'Mozilla/5.0'}
        html = requests.get(url, headers=headers)
        # url에 정보를 요청
        soup = BeautifulSoup(html.text, 'html.parser') # bs4 파씽

        # html 확인/ 필요한 정보가 들어있는 태그 추출
        total = soup.find_all('div','user_reviews_filter_menu_flyout_content')

        for i in total:
            positive_txt = i.text.split('\n\n')[2]
            break

        Pnum = positive_txt.lstrip('Positive\xa0(')
        positive_num = Pnum.rstrip(')')
        positive_num = positive_num.replace(',', '')

        return int(positive_num)

    except:
        return 0
```

```python
import requests # 파이썬 기본 패키지
from bs4 import BeautifulSoup      # bs4 로드

def find_negative(id):
    try:
        url = f"https://store.steampowered.com/app/{id}/"

        headers = {'User-Agent': 'Mozilla/5.0'}
        html = requests.get(url, headers=headers)
        # url에 정보를 요청
        soup = BeautifulSoup(html.text, 'html.parser') # bs4 파씽

        # html 확인/ 필요한 정보가 들어있는 태그 추출
        total = soup.find_all('div','user_reviews_filter_menu_flyout_content')

        for i in total:
            negative_txt = i.text.split('\n\n')[3]
            break


        Nnum = negative_txt.lstrip('Negative\xa0(')
        negative_num = Nnum.rstrip(')\n')
        negative_num = negative_num.replace(',', '')

        return int(negative_num)

    except:
        return 0
```

#### 긍정적 평가수와 개발사/출판사 간의 통계적 의미 분석

##### 개발사와 긍정적 평가수간의 통계적 의미 분석

- Developers 열은 범주형 변수열이므로 레이블 인코딩을 통해 숫자로 변환해 'Developers_cat'열에 저장한다.
- 귀무가설: 개발사와 긍정적 평가수간은 통계적으로 의미가 없을 것이다.

```python
df_steam_2 = df_steam_1.copy()
```

```python
df_steam_2['Developers_cat'] = df_steam_2['Developers'].astype('category').cat.codes
```

```python
from scipy.stats import *

shapiro(df_steam_2['Developers_cat'])
```

- 'Developers_cat'은 p-value가 6.965539774887635e-89로 정규성을 만족하지 않는다.
- 그러므로, 스피어만 상관계수를 구한다.

```python
spearmanr(df_steam_2['Developers_cat'], df_steam_2['Positive'])
```

- p-value가 0.0002이므로 귀무가설 기각. 따라서, 개발사와 긍정적 평가수간은 통계적으로 유의미하다.
- 하지만 계수가 -0.01로 긴밀한 관계는 아님을 알 수 있다.

##### 배급사와 긍정적 평가수간의 통계적 의미 분석

- Publishers 열은 범주형 변수열이므로 레이블 인코딩을 통해 숫자로 변환해 'Publishers_cat'열에 저장한다.
- 귀무가설: 배급사와 긍정적 평가수간은 통계적으로 유의미하지 않을 것이다.

```python
df_steam_2['Publishers_cat'] = df_steam_2['Publishers'].astype('category').cat.codes
```

```python
shapiro(df_steam_2['Publishers_cat'])
```

- 'Publishers_cat'은 p-value가 7.868923831557018e-89로 정규성을 만족하지 않는다.
- 그러므로, 스피어만 상관계수를 구한다.

```python
spearmanr(df_steam_2['Publishers_cat'], df_steam_2['Positive'])
```

- p-value가 0.003이므로 귀무가설 기각. 따라서, 배급사와 긍정적 평가수간은 통계적으로 유의미하다.
- 하지만, 계수가 -0.009 이므로 긴밀한 관계는 아님을 알 수 있다.

####스팀데이터 분석 정리 및 결론

- 게임을 많이 만든 개발사나 배급사라고 해서 Positive가 높은 게임을 만드는 것은 아님을 알 수 있다.
- 동일 시리즈를 많이 냈다고 해서 Positive가 높은 게임은 아니다.
- Positive가 높은 게임이 Negative가 높은 것으로 보아 유저 수가 많아 그 안에서 호불호가 갈린 다는 것을 알 수 있으므로 유저를 많이 모으는 게 관건임을 추측할 수 있다.
- 스팀데이터에서 상관관계를 분석한 결과, Positive와 긴밀한 상관관계를 보이는 것은 없었다.
- 유저들이 유료/무료 여부에 상관이 없다.
- 언어는 최대한 많이 지원해야 한다.
- 카테고리와 장르가 Positive와 상관이 없었다. 하지만, 장르에서 Casual이 상위 데이터에 적었고 Action이 상위 데이터에 많이 있었던 것으로 보아 장르들 중 Action 장르가 유저를 많이 유입할 수 있는 가능성이 가장 높다고 추측할 수 있다.
- 개발사와 배급사는 Positive와 관계는 있지만, 긴밀한 관계가 아니다.
<br><br>
##### 결론
따라서, 본 프로젝트의 소규모 게임개발 회사 포지션 입장으로서 개발사와 배급사는 Positive와 관계는 있지만 긴말하지는 않으므로 충분히 흥행 게임을 만들 가능성이 있다.<br><br>
- 유료/무료 여부가 Positive와 상관이 없으므로 유료 게임을 만들더라도 언어를 최대한 많이 지원해야 하고 Action 장르로 개발하는 것이 유저를 모으는데 효과적이라고 추측할 수 있다.
- DLC 수보단 여부가 더 관련이 있었으므로 DLC가 있는 것이 효과적일 것이다.

```python
# 표준화 (StandardScaler)
X_features = df_steam_3[selected_col]
X_features_scaled = StandardScaler().fit_transform(X_features)
```

```python
### 여러개의 클러스터링 갯수를 List로 입력 받아 각각의 실루엣 계수를 면적으로 시각화한 함수 작성
def visualize_silhouette(cluster_lists, X_features):
    from sklearn.cluster import KMeans
    from sklearn.metrics import silhouette_samples, silhouette_score

    import matplotlib.pyplot as plt
    import matplotlib.cm as cm
    import numpy as np
    import math

    n_clusters = len(cluster_lists)

    n_cols = min(n_clusters, 5)
    n_rows = math.ceil(n_clusters / n_cols)

    fig, axs = plt.subplots(figsize=(4 * n_cols, 4 * n_rows), nrows=n_rows, ncols=n_cols)

    axs = np.array(axs).reshape(-1) if n_clusters > 1 else [axs]

    for ind, n_cluster in enumerate(cluster_lists):

        clusterer = KMeans(n_clusters=n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(X_features)

        sil_avg = silhouette_score(X_features, cluster_labels)
        sil_values = silhouette_samples(X_features, cluster_labels)

        y_lower = 10
        axs[ind].set_title(f'Number of Clusters: {n_cluster}\nSilhouette Score: {round(sil_avg, 3)}')
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([0, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])

        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[cluster_labels == i]
            ith_cluster_sil_values.sort()

            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values,
                                   facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10

        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")

    for ax in axs[n_clusters:]:
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def visualize_kmeans_plot_multi(cluster_lists, X_features):
    from sklearn.cluster import KMeans
    from sklearn.decomposition import PCA
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import math

    n_clusters = len(cluster_lists)
    n_cols = min(n_clusters, 5)
    n_rows = math.ceil(n_clusters / n_cols)

    fig, axs = plt.subplots(figsize=(4 * n_cols, 4 * n_rows), nrows=n_rows, ncols=n_cols)

    axs = np.array(axs).reshape(-1) if n_clusters > 1 else [axs]

    pca = PCA(n_components=2)
    pca_transformed = pca.fit_transform(X_features)
    dataframe = pd.DataFrame(pca_transformed, columns=['PCA1', 'PCA2'])

    for ind, n_cluster in enumerate(cluster_lists):

        clusterer = KMeans(n_clusters=n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(pca_transformed)
        dataframe['cluster'] = cluster_labels

        unique_labels = np.unique(cluster_labels)
        markers = ['o', 's', '^', 'x', '*', 'D', 'v', 'p', 'H', '+']

        for label in unique_labels:
            label_df = dataframe[dataframe['cluster'] == label]
            cluster_legend = f'Cluster {label}' if label != -1 else 'Noise'
            axs[ind].scatter(x=label_df['PCA1'], y=label_df['PCA2'], s=70,
                             edgecolor='k', marker=markers[label % len(markers)], label=cluster_legend)

        axs[ind].set_title(f'Number of Clusters: {n_cluster}')
        axs[ind].legend(loc='upper right')

    for ax in axs[n_clusters:]:
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```

#### 하이퍼파라미터 튜닝

```python
from hyperopt import hp
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.metrics import roc_auc_score
from hyperopt import fmin, tpe, Trials
import numpy as np

# 1단계 공간탐색
xgb_search_space = {'max_depth': hp.quniform('max_depth', 5, 15, 1),
                    'min_child_weight': hp.quniform('min_child_weight', 1, 6, 1),
                    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 0.95),
                    'learning_rate': hp.uniform('learning_rate', 0.01, 0.2)}

# 2단계 목적함수 (h.p에 따른 손실값을 구함.)
def objective_func(search_space):
    xgb_clf = XGBClassifier(n_estimators=100, max_depth=int(search_space['max_depth']),
                            min_child_weight=int(search_space['min_child_weight']),
                            colsample_bytree=search_space['colsample_bytree'],
                            learning_rate=search_space['learning_rate'],
                            early_stopping_rounds=30, eval_metric='auc',
                            objective='multi:softprob', # To ensure multi-class objective
                            num_class=len(y_train.unique()) # Number of classes
                           )
    # 3개 k-fold 방식으로 평가된 roc_auc 지표를 담는 list
    roc_auc_list= []

    # 3개 k-fold방식 적용
    kf = StratifiedKFold(n_splits=3)
    # X_train을 다시 학습과 검증용 데이터로 분리

    for tr_index, val_index in kf.split(X_train, y_train):
        # kf.split(X_train)으로 추출된 학습과 검증 index값으로 학습과 검증 데이터 세트 분리
        X_tr, y_tr = X_train.iloc[tr_index], y_train.iloc[tr_index]
        X_val, y_val = X_train.iloc[val_index], y_train.iloc[val_index]
        # early stopping은 30회로 설정하고 추출된 학습과 검증 데이터로 XGBClassifier 학습 수행.
        xgb_clf.fit(X_tr, y_tr,
                   eval_set=[(X_tr, y_tr), (X_val, y_val)])

        # Calculate ROC AUC for multi-class using 'ovo' or 'ovr'
        score = roc_auc_score(y_val, xgb_clf.predict_proba(X_val), multi_class='ovo') # or multi_class='ovr'
        roc_auc_list.append(score)

    # 3개 k-fold로 계산된 roc_auc값의 평균값을 반환하되,
    # HyperOpt는 목적함수의 최소값을 위한 입력값을 찾으므로 -1을 곱한 뒤 반환.
    return -1 * np.mean(roc_auc_list)
```

```python
trials = Trials()

# fmin()함수를 호출. max_evals지정된 횟수만큼 반복 후 목적함수의 최소값을 가지는 최적 입력값 추출.
best = fmin(fn=objective_func,  # 목적함수명
            space=xgb_search_space,  #위에서 설정한 검색공간
            algo=tpe.suggest, #알고리즘 인자의 기본값
            max_evals=5, # 최대 반복 횟수
            trials=trials, rstate=np.random.default_rng(seed=30)) #임의의 랜덤시드값

print('best:', best)
```

#### 모델 개선

```python
xgb_clf2 = XGBClassifier(n_estimators=500, learning_rate=round(best['learning_rate'], 5),
                        max_depth=int(best['max_depth']), min_child_weight=int(best['min_child_weight']),
                        colsample_bytree=round(best['colsample_bytree'], 5),
                        random_state=156, early_stopping_rounds=100,eval_metric="auc"
                       )

xgb_clf2.fit(X_train, y_train,eval_set=[(X_train, y_train), (X_test, y_test)])

# Binarize the labels for multi-class ROC AUC calculation
lb = LabelBinarizer()
lb.fit(y_test)
y_test_binarized2 = lb.transform(y_test)

xgb_roc_score2 = roc_auc_score(y_test_binarized2, xgb_clf.predict_proba(X_test), multi_class='ovo') # predict_proba is used without [:,1]
#                            검증피쳐       양성 클래스 확률
print(f'ROC AUC: {round(xgb_roc_score2,4)}')
```
